### What is a Variable?

A variable is a container used to store data in memory.
In Python, variables are created when a value is assigned to them.

Variables help us store and reuse information in a program.

### Creating Variables in Python

In Python, we create a variable using the assignment operator (=).

The value on the right side is stored in the variable on the left side.

```python
name = "Muskan"
age = 26
salary = 25000
is_working = True

print(name)
print(age)
print(type(age))
```

### Data Types in Variables

- Integer (int) → Whole numbers (e.g., 10, 25)
- Float → Decimal numbers (e.g., 5.5, 99.9)
- String (str) → Text values (e.g., "Muskan")
- Boolean (bool) → True or False values

### Dynamic Typing in Python

Python is dynamically typed.
This means the type of a variable can change during execution.

```python
x = 10
print(type(x))

x = "Data Science"
print(type(x))
```
