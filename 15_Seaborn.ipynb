# Introduction to Seaborn

Seaborn is a Python library used for statistical data visualization.  
It is built on top of Matplotlib and provides more attractive and informative graphs.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Creating a Dataset

We create a sample dataset using Pandas to visualize data using Seaborn.

```python
data = {
 "Name": ["Ali","Sara","John","Aman","Riya"],
 "Age": [22,24,23,26,21],
 "Salary": [30000,40000,35000,50000,28000],
 "City": ["Mumbai","Delhi","Pune","Mumbai","Delhi"]
}

df = pd.DataFrame(data)
print(df)
```

# Scatter Plot

A scatter plot shows the relationship between two numerical variables.  
Each point represents one observation in the dataset.

```python
sns.scatterplot(x="Age", y="Salary", data=df)

plt.title("Age vs Salary")

plt.show()
```

# Line Plot

A line plot shows trends over a continuous variable.  
It connects data points using lines.

```python

sns.lineplot(x="Age", y="Salary", data=df)

plt.title("Salary Trend")

plt.show()
```

# Bar Plot

A bar plot is used to compare values across different categories.  
It is commonly used to visualize aggregated data.

sns.barplot(x="City", y="Salary", data=df)

plt.title("Average Salary by City")

plt.show()


```python
sns.barplot(x="City", y="Salary", data=df)

plt.title("Average Salary by City")

plt.show()
```


# Count Plot

A count plot shows the frequency of each category in the dataset.  
It is useful for understanding categorical data distribution.

```python
sns.countplot(x="City", data=df)

plt.title("City Distribution")

plt.show()
```

# Histogram

A histogram shows the distribution of a numerical variable.  
It groups values into ranges and shows their frequency.

```python
sns.histplot(df["Age"])

plt.title("Age Distribution")

plt.show()
```

# Box Plot

A box plot shows the distribution of data and helps detect outliers.  
It displays median, quartiles, and spread of the data.

```python
sns.boxplot(x="City", y="Salary", data=df)

plt.title("Salary Distribution by City")

plt.show()
```

# Heatmap

A heatmap visualizes correlations between numerical variables.  
It uses color intensity to represent values.

```python
sns.heatmap(df.corr(), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

# Pair Plot

A pairplot shows pairwise relationships between numerical variables.  
It is useful for exploring patterns in the dataset.

```python
sns.pairplot(df)

plt.show()
```
