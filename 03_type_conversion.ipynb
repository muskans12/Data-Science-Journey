Type conversion means converting one data type into another data type.

Python provides built-in functions like int(), float(), str(), and bool() to perform type conversion.

## Integer Conversion

The int() function is used to convert a value into an integer.

```python
x = "10"

y = int(x)

print(y)
print(type(y))
```

## Float Conversion

The float() function converts a value into a decimal number.

```python
a = 5

b = float(a)

print(b)
print(type(b))
```

## String Conversion

The str() function converts numbers into text format.

```python
age = 26

text = str(age)

print(text)
print(type(text))
```

## Boolean Conversion

The bool() function converts values into True or False.

```python
print(bool(0))
print(bool(10))
```

## Example of Type Conversion

Sometimes we need to convert data types before performing operations.

```python
x = "50"
y = 20

x = int(x)

result = x + y

print(result)
```
