# Python Strings

A string is a sequence of characters.

Strings in Python are written using:
- Single quotes (' ')
- Double quotes (" ")

Example:
"Hello"
'Python'

## 1. Creating a String

```python
name = "Python"
print(name)
```

## 2. Accessing Characters in a String

We can access characters in a string using indexing.
Indexing starts from 0.

```python
text = "Python"

print(text[0])
print(text[1])
print(text[2])
```

## 3. String Slicing

Slicing allows us to extract a part of a string.

```python
text = "Python Programming"

print(text[0:6])
print(text[7:18])
```

## 4. String Length

```python
text = "Data Science"

print(len(text))
```

## 5. Convert String Case

```python
text = "python"

print(text.upper())
print(text.lower())
```

## 6. Remove Spaces

```python
text = "   Python   "

print(text.strip())
```

## 7. Replace String

```python
text = "I like Java"

print(text.replace("Java","Python"))
```

## 8. Split String

```python
text = "Python Data Science"

print(text.split())
```

## 9. String Concatenation

```python
first = "Data"
second = "Science"

print(first + " " + second)
```
