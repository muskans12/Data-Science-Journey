# Introduction to Pandas

Pandas is a Python library used for data analysis and data manipulation.  
It provides powerful data structures like Series and DataFrame to work with structured data.

```python
import pandas as pd
```

# Pandas Series

A Series is a one-dimensional labeled array capable of holding any data type.  
It is similar to a column in a table.

```python
data = pd.Series([10,20,30,40])
print(data)
```

# Creating a DataFrame

A DataFrame is a two-dimensional data structure that stores data in rows and columns.  
It is similar to a table in Excel or SQL.

```python
data = {
 "Name": ["Ali","Sara","John"],
 "Age": [22,24,23],
 "City": ["Mumbai","Delhi","Pune"]
}

df = pd.DataFrame(data)
print(df)
```

# Viewing Data

Pandas provides functions to quickly view the dataset.  
These functions help understand the structure of the data.

```python
df.head()
df.tail()
```

# DataFrame Information

The info() function shows details about the dataset such as column names, data types, and missing values.

```python
df.info()
```

# Shape of DataFrame

The shape attribute returns the number of rows and columns in the dataset.

```python
df.shape
```

# Selecting Columns

You can select a specific column from a DataFrame using its column name.

```python
df["Name"]
```

```python
df[["Name","Age"]]
```

# Row Selection using loc

The loc function is used to select rows and columns by labels.

```python
df.loc[0]
```

```python
df.loc[1,"Name"]
```

# Row Selection using iloc

The iloc function is used to select rows and columns using index positions.

```python
df.iloc[0]
```

```python
df.iloc[1,0]
```

# Filtering Data

Filtering helps select rows based on specific conditions.

```python
df[df["Age"] > 22]
```

# GroupBy Operation

GroupBy is used to group data based on a column and perform aggregation operations.

```python
df.groupby("City")["Age"].mean()
```

# Handling Missing Values

Missing values are common in real-world datasets.  
Pandas provides functions to detect and handle missing data.

```python
df.isnull()
df.fillna(0)
df.dropna()
```

# Statistical Functions

Pandas provides functions to calculate statistical measures like mean, sum, max, and min.

```python
df["Age"].mean()
df["Age"].max()
df["Age"].min()
```
