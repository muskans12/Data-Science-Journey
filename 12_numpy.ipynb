# Introduction to NumPy

NumPy is a Python library used for numerical computing.  
It helps perform fast mathematical operations on arrays.

```python
import numpy as np
```

# Creating NumPy Array

NumPy arrays allow storing multiple values in a single variable.  
They are faster and more efficient than Python lists.

```python
arr = np.array([10,20,30,40,50])
print(arr)
```

# Creating 2D Array

A 2D array represents data in rows and columns like a table.  
It is commonly used in data analysis and machine learning.

```python
arr2 = np.array([[1,2,3],
                 [4,5,6]])

print(arr2)
```

# Array Properties

NumPy provides attributes to understand the structure of an array.  
These include shape, size, number of dimensions, and data type.

```python
print(arr.shape)
print(arr.size)
print(arr.ndim)
print(arr.dtype)
```

# Special Arrays

NumPy can create arrays filled with zeros or ones quickly.  
These arrays are useful for initializing data structures.

```python
print(np.zeros(5))
print(np.ones(5))
```

# Range Arrays

NumPy provides functions to generate numbers in a range.  
`arange()` and `linspace()` create evenly spaced values.

```python
print(np.arange(1,10))
print(np.linspace(1,10,5))
```

# Indexing

Indexing is used to access a specific element from an array.  
Index numbers start from 0 in Python.

```python
print(arr[0])
print(arr[2])
```

# Slicing

Slicing is used to extract a portion of the array.  
The format used is `[start:end]`.

```python
print(arr[1:3])
```

# Step Slicing

Step slicing allows skipping elements while selecting data.  
The format used is `[start:end:step]`.

```python
print(arr[0:5:2])
```

# Reshape

Reshape changes the structure of an array without changing the data.  
It converts a 1D array into rows and columns.

```python
arr3 = np.array([1,2,3,4,5,6])
print(arr3.reshape(2,3))
```

# Mathematical Operations

NumPy performs element-wise mathematical operations.  
Operations apply to every element in the array.

```python
arr = np.array([10,20,30])

print(arr + 5)
print(arr * 2)
```

# Statistical Functions

NumPy provides functions to analyze numerical data.  
Common functions include sum, mean, maximum, and minimum.

```python
arr = np.array([10,20,30,40,50])

print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
```
