# Python Lists

A list is a collection of items stored in a single variable.

Lists are:
- Ordered
- Changeable (mutable)
- Allow duplicate values

Example:
[10, 20, 30]
["apple", "banana", "mango"]

## 1. Creating a List

```python
numbers = [10, 20, 30, 40]

print(numbers)
```

## 2. Access List Elements

List elements are accessed using indexing.

Index starts from 0.

```python
fruits = ["apple", "banana", "mango"]

print(fruits[0])
print(fruits[1])
print(fruits[2])
```

## 3. Negative Indexing

```python
fruits = ["apple", "banana", "mango"]

print(fruits[-1])
print(fruits[-2])
```

## 4. List Slicing

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])
```

## 5. Change List Items

```python
fruits = ["apple", "banana", "mango"]

fruits[1] = "orange"

print(fruits)
```

## 6. Add Items to List

```python
fruits = ["apple", "banana"]

fruits.append("mango")

print(fruits)
```

## 7. Insert Item

```python
fruits = ["apple", "banana"]

fruits.insert(1, "mango")

print(fruits)
```

## 8. Remove Items

```python
fruits = ["apple", "banana", "mango"]

fruits.remove("banana")

print(fruits)
```

## 9. List Length

```python
numbers = [10, 20, 30, 40]

print(len(numbers))
```

## 10. Loop Through List

```python
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```
