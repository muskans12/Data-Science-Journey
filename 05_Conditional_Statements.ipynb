# Python Conditional Statements (if, elif, else)

Conditional statements are used to perform different actions based on different conditions.

They allow a program to make decisions.

Examples:
- If age >= 18 → Eligible for voting
- If marks >= 40 → Pass

## 1. if Statement

The `if` statement executes a block of code only when the given condition is True.

```python
age = 20

if age >= 18:
    print("You are eligible to vote")
```

## 2. if else Statement

The `if else` statement executes one block of code when the condition is True and another block when the condition is False.

```python
age = 16

if age >= 18:
    print("Eligible for voting")
else:
    print("Not eligible for voting")
```

## 3. if elif else Statement

The `elif` statement is used to check multiple conditions.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

## 4. Nested if Statement

A nested `if` statement means using an `if` statement inside another `if`.

```python
age = 22
citizenship = "India"

if age >= 18:
    if citizenship == "India":
        print("Eligible to vote")
```

# 5. Logical Operators with Conditions

Logical operators are used to combine multiple conditions.

Types:
- and
- or
- not

```python
age = 25
salary = 40000

if age > 21 and salary > 30000:
    print("Loan Approved")
```
