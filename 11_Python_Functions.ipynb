# Python Functions

A function is a block of code that runs when it is called.

Functions help to:
- reuse code
- make programs organized
- reduce repetition

## 1. Creating a Function

A function is created using the `def` keyword.

```python
def greet():
    print("Hello, welcome to Python")

greet()
```

## 2. Function with Parameters

Parameters allow us to pass information into a function.

```python
def greet(name):
    print("Hello", name)

greet("Muskan")
```

## 3. Function with Multiple Parameters

```python
def add(a, b):
    print(a + b)

add(10, 20)
```

## 4. Function with Return Value

The `return` keyword is used to send a result back.

```python
def multiply(a, b):
    return a * b

result = multiply(5, 4)

print(result)
```

## 5. Default Parameters

```python
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Muskan")
```

## 6. Lambda Function

A lambda function is a small anonymous function.

```python
square = lambda x: x * x

print(square(5))
```

## 7. Function with List

```python
def total(numbers):
    return sum(numbers)

data = [10, 20, 30]

print(total(data))
```
