# Python Dictionaries

A dictionary is a collection of data stored in **key-value pairs**.

Dictionaries are:
- Ordered
- Changeable
- Do not allow duplicate keys

Example:
{
"name": "John",
"age": 25,
"city": "New York"
}

## 1. Creating a Dictionary

```python
person = {
    "name": "John",
    "age": 25,
    "city": "New York"
}

print(person)
```

## 2. Access Dictionary Values

```python
person = {
    "name": "John",
    "age": 25,
    "city": "New York"
}

print(person["name"])
print(person["age"])
```

## 3. Using get() Method

```python
person = {
    "name": "John",
    "age": 25
}

print(person.get("name"))
```

## 4. Adding Items to Dictionary

```python
person = {
    "name": "John",
    "age": 25
}

person["city"] = "London"

print(person)
```

## 5. Updating Dictionary

```python
person = {
    "name": "John",
    "age": 25
}

person["age"] = 30

print(person)
```

## 6. Removing Items

```python
person = {
    "name": "John",
    "age": 25,
    "city": "Paris"
}

person.pop("city")

print(person)
```

## 7. Dictionary Keys and Values

```python
person = {
    "name": "John",
    "age": 25,
    "city": "Paris"
}

print(person.keys())
print(person.values())
```

## 8. Loop Through Dictionary

```python
person = {
    "name": "John",
    "age": 25,
    "city": "Paris"
}

for key, value in person.items():
    print(key, value)
```
