# Introduction to Matplotlib

Matplotlib is a Python library used for data visualization.  
It allows us to create charts and graphs to understand data easily.

```python
import matplotlib.pyplot as plt
```

# Line Plot

A line plot is used to show trends over time or continuous data.  
It connects data points using a straight line.

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y)
plt.show()
```

# Adding Title and Labels

Titles and labels help make the chart more understandable.  
They describe what the axes and graph represent.

```python
plt.plot(x,y)

plt.title("Sales Growth")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

# Adding Grid

A grid makes the chart easier to read by showing reference lines.

```python
plt.plot(x,y)

plt.title("Sales Growth")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid()

plt.show()
```

# Bar Chart

A bar chart is used to compare different categories.  
Each category is represented by a rectangular bar.

```python
products = ["Laptop","Phone","Tablet"]
sales = [50,120,80]

plt.bar(products,sales)

plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")

plt.show()
```

# Histogram

A histogram shows the distribution of numerical data.  
It groups values into bins and displays their frequency.


```python
data = [10,20,20,30,30,30,40,40,50]

plt.hist(data)

plt.title("Data Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

# Scatter Plot

A scatter plot shows the relationship between two numerical variables.  
Each point represents a pair of values.


```python
height = [150,160,170,180]
weight = [50,60,65,75]

plt.scatter(height,weight)

plt.title("Height vs Weight")
plt.xlabel("Height")
plt.ylabel("Weight")

plt.show()
```

# Pie Chart

A pie chart shows the percentage share of different categories.  
Each slice represents a part of the whole.

```python
values = [50,30,20]
labels = ["Laptop","Phone","Tablet"]

plt.pie(values, labels=labels)

plt.title("Sales Share")

plt.show()
```

# Pie Chart with Percentage

The autopct parameter displays percentage values on the pie chart.

```python
plt.pie(values, labels=labels, autopct='%1.1f%%')

plt.title("Sales Share")

plt.show()
```
