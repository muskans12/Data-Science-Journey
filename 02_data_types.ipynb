Data types represent the type of value stored in a variable.

Python automatically detects the data type based on the value assigned to the variable.

```python
age = 26
price = 50.5
name = "Muskan"
is_working = True

print(type(age))
print(type(price))
print(type(name))
print(type(is_working))
```

## Types of Data Types

Integer → Whole numbers  
Float → Decimal numbers  
String → Text values  
Boolean → True or False

```python
city = "Pune"
gym_days = 5
salary = 25000.75
learning_python = True

print(type(city))
print(type(gym_days))
print(type(salary))
print(type(learning_python))
```
